```python
import numpy as np
import pandas as pd

df_global = pd.read_csv('global_plastics_production.csv')

print(df_global.head())
```

```python
df_global.drop(['Entity'], axis=1, inplace=True)
 
df_global.rename(columns={'Global plastics production (million tonnes) (tonnes)':'Production' }, inplace=True)
 
df_global.columns = list(map(str, df_global.columns))

print(df_global.head()) 
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
 
print(plt.style.available)

mpl.style.use(['ggplot']) 
```

```python
df_global.plot.line(x = 'Year', y = 'Production', color = 'blue', fontsize = 20, figsize = (20,10), title='Global Plastic Production, 2015')

plt.ylabel('Billion Tonnes', fontsize = 20)

plt.xlabel('Year', fontsize = 20)

plt.plot(1965, 17000000, 'gp',  markersize=15)
plt.plot(1988, 110000000, 'gp', markersize=15)

plt.text(1966, 5000000, '1965 Commercial Plastic Bags', color = 'black', fontsize = 20)

plt.text(1989, 110000000, '1988 Australia Polymer Bank Notes', color = 'black', fontsize = 20 )
plt.savefig('Global Plastic Production 2015.jpg' ,bbox_inches = 'tight')
plt.show()
 
```

```python
df_production_waste_per_sector = pd.read_csv('plastic_production_waste_by_sector.csv',index_col=0)

print(df_production_waste_per_sector.head(3))
```

```python
df_production_waste_per_sector.rename(columns = {'Waste Generated':'Waste'}, inplace=True)

df_production_waste_per_sector.drop(columns = ['Code', 'Year'], inplace=True)

print(df_production_waste_per_sector)
```

```python
df_production_waste_per_sector.plot(kind = 'barh', width = 0.8, fontsize = 10, figsize = (20,10),  layout = 'tight', title = 'Plastic Production vs Waste, 2015')

plt.xlabel('Billion Tonnes')

plt.ylabel(' ')

plt.savefig('Plastic Production vs Waste 2015.jpg' ,bbox_inches = 'tight')

plt.show()
```
